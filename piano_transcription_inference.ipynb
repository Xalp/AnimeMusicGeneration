```python
import os
path=[]
for dirname, _, filenames in os.walk('./mp3s_piano_solo/'):
    for filename in filenames:
        path.append(os.path.join(dirname, filename))
```

```python
from piano_transcription_inference import PianoTranscription, sample_rate, load_audio

# Transcriptor
transcriptor = PianoTranscription(device='cuda', checkpoint_path='CRNN_note_F1=0.9677_pedal_F1=0.9186.pth')

for each in path:
    # Load audio
    (audio, _) = load_audio(each, sr=sample_rate, mono=True)


    # Transcribe and write out to MIDI file
    transcribed_dict = transcriptor.transcribe(audio, each[:-3]+"mid")
```
